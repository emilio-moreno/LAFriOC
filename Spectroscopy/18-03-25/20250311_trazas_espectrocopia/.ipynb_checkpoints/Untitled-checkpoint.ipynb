```python
import pandas as pd
import matplotlib.pyplot as plt
import os 

listdir = os.listdir()
```

```python
for filename in listdir:
    if not ".CSV" in filename: continue
    df1 = pd.read_csv(filename, names=["1", "2", "3", "freq", "amp", "6"])
    plt.plot(df1["freq"], df1["amp"], linewidth=0.5)
    plt.show()
```
