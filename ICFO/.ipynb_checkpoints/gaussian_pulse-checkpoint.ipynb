```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 40
k = 4
A = 3
b = 0.1
x = np.linspace(-a, a, 10**5)
color = "#b01b1f"

f = lambda x: np.cos(k*x)
g = lambda x: A*np.exp(-b*x**2)


fig, ax = plt.subplots(dpi=1000)
ax.plot(x, f(x)*g(x), c=color)
ax.set_aspect('equal')
ax.set_ylim(-A - 1, A + 1)

plt.savefig("gaussianPulse_red.png")
plt.show()
```
