```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 3, 3, 2
b_step = 0.25
z0 = 1
omega_0 = 1

z = np.linspace(-a, a, 10**3)
Z1 = np.arange(0, b, b_step)
omega = lambda z: omega_0 * np.sqrt(1 + (z/z0)**2)
R = lambda z: z * (1 + (z0/z)**2)

w_color = "#365393"
R_color = "#554c48"

fig, ax = plt.subplots(dpi=100)
ax.plot(z, omega(z), c=w_color)
ax.plot(z, -omega(z), c=w_color)

for z1 in Z1:
    r = np.linspace(-omega(z1), omega(z1), 10**3)
    ax.plot(z1+np.sqrt(R(z1)**2 - r**2), r, label=f"{z1}")#, c=R_color)
    ax.plot(-z1-np.sqrt(R(z1)**2 - r**2), r, label=f"{z1}")#, c=R_color)
plt.legend()
```
